# Helpers methods

```python
def executeNormalTest(dataframe):
    for item in dataframe:
        statistics_test_generic, p_value_generic = normaltest(dataframe[item])
        print(f'{item} not a normal distribution\n' if p_value_generic <= SIGNIFICANCE else f'{item} is a normal distribution \n')
    
def createHistogram(dataframe):
    figure, ax = plt.subplots(figsize = (15, 15))
    _ = dataframe.hist(ax=ax, bins=15)

def createBoxplots(dataframe, column):
    sns.boxplot(x = column, data = dataframe, orient = 'h')

def get_numeric_variables(dataframe):
    numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
    return dataframe.select_dtypes(include=numerics)
```

### Race by original language and release date(year) - curiosity moment

```python
dataframe_race_plot = data.copy()
dataframe_race_plot.drop('budget', axis=1, inplace=True)
dataframe_race_plot.drop('original_title', axis=1, inplace=True)
dataframe_race_plot.drop('production_companies', axis=1, inplace=True)
dataframe_race_plot.drop('Keywords', axis=1, inplace=True)
dataframe_race_plot.drop('cast', axis=1, inplace=True)
dataframe_race_plot.drop('crew', axis=1, inplace=True)
dataframe_race_plot.drop('revenue', axis=1, inplace=True)
dataframe_race_plot.drop('main_genre', axis=1, inplace=True)
dataframe_race_plot.drop('num_genres', axis=1, inplace=True)
dataframe_race_plot.drop('num_production_countries', axis=1, inplace=True)
dataframe_race_plot.drop('main_production_country', axis=1, inplace=True)
dataframe_race_plot.drop('runtime', axis=1, inplace=True)
for i in set(dataframe_race_plot.original_language):
    dataframe_race_plot[i] = 0

sum_languages = dict.fromkeys(set(dataframe_race_plot.original_language), 0)
dataframe_race_plot['release_date'] = pd.to_datetime(dataframe_race_plot['release_date'], errors='coerce')
dataframe_race_plot = dataframe_race_plot.sort_values(by="release_date")
dataframe_race_plot['year'] = dataframe_race_plot.release_date.dt.year

for i, row in dataframe_race_plot.iterrows():
    original_language = row['original_language']
    sum_languages[original_language] = sum_languages[original_language] + 1
    dataframe_race_plot.at[i, original_language] = sum_languages[original_language]
    for item in set(dataframe_race_plot.original_language):
        if item != original_language:
            dataframe_race_plot.at[i, item] = sum_languages[item]
    
dataframe_race_plot.drop('original_language', axis=1, inplace=True)
dataframe_race_plot.drop('release_date', axis=1, inplace=True)

dataframe_race_plot.reset_index(drop=True, inplace=True)
dataframe_race_plot.set_index('year', inplace=True)

dataframe_race_plot = dataframe_race_plot.groupby('year')
dataframe_race_plot.head()
```
